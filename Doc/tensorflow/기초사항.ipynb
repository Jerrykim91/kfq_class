#### 인공지능, 인공신경망

- 인공 신경망은 인공지능을 구현하는 알고리즘 이다
- 두뇌를 구성하는 뉴런(신경세포)를 본따서 시스템 구성
- 신경망을 이용하여서 지능을 발현하는 원리를 기계에 적용한 방식

#### history (발전->암흑->발전->암흑->발전...)

- 1956년 존 매카시 주장해서 처음으로 등장
- 1980년 전문분야의 지식을 모은 전문가 시스템으로 다시한번 도약
- 1990년 인터넷, 컴퓨터 시스템의 발전 -> 역전파 알고리즘 등장 
- 1997년 IBM 딥블루 -> 체스 세계 챔피언에 대전에서 승리
- 2006년 토렌토 대학, 제프리 힌트 교수 => 합성곱(다계층)에 대한 인공신경망의 효율적 학습 가능성에 대한 논문 등장 => 딥러닝이라고 부른다!!
- 2011년 퀴드대회 제프디 => 인공지능 왓슨 우승 => 의료, 금융분야 확대
- 2012년 이미지 인식 대회 딥러닝으로 출전한 제프리 힌트 팀이 딥러닝으로 추천. 2014년 구글 우승, 그전까지 대부분 참가자들은 머신러닝으로 경쟁했다면, 이후로는 딥러닝으로 경쟁하는 패더라임 변경 => 인식율 90% 달성
- 2016년 강화학습이라는 방법으로 알파고가 바둑에서 이세돌에게 승리 -> 전세계적인 이슈를 달성 
- GPU가 효율적인 딥러닝 학습이 가능하다는것이 알려지게 되면서, 한대의 PC 에서도 
처리가 가능한 구조로 변경 => 블록체인의 채굴!!
- 4차 산업의 핵심 기술로 자리를 잡았다

#### 딥러닝

- Deep Learing
- 여러 계층(layer)를 가진 신경망(Neural Network)을 사용하여 머신러닝을 수행
- 머신러닝의 한 범주
- 주목받는 시점
 > 2012년 이미지 인식 대회 ILSVRC에서 주목 (영상인식/분류 소프트웨어 대회)    
 > 제프리 힌튼 교수팀 딥러닝 기법으로 처음 등장  
 > 2013년 부터는 상위권이 거의 다 딥러닝, 2014년 구글이 우승  
 
- 딥러닝은 이미지(영상)인식, 음석인식, 자연어처리 분야에 큰 성과를 내고 있다

#### 이미지 인식의 차이점

- **특징량 추출**이라 부분이 가장 큰 차이가 발생
- 기존법 : 사람이 이미지의 내부 특징을 하나 하나 체크해서 잡아줘야 한다
- 딥러닝 : 학습 데이터에서 기계가 자동으로 특징을 추출한다 -> 스스로 특징을 추출해서 학습

#### DNN(Deep Neural Network)

- 뉴럴 네트워크 : 인간의 신경망을 본따서 만든 네트워크 구조
 > 인간의 뇌에 잇는 신경세포 : 뉴런  
 > 하나의 뉴런이 다른 뉴런에게 신호를 받고, 전달하는 역확을 담당  
 > 신경의 흐름을 기반으로 다양한 정보를 생성해 낸다  
- 신경망 
 > 이런 뉴런들의 연결되어 있는 구조의 망  
- 신경망이 3개 이상 중첩되면 => DNN
 > 이런 신경망을 이용하여 기계학습을 수행하면 딥러닝이고 한다  
 > 대량의 데이터로 학습  
 > 학습 => 뉴런의 접속 가중치를 조정하는 행위  

#### 퍼셉트론  (화요일)

- 기계 학습의 기초가 되는 핵심 개념
- 프랑크 로젠블라트 1957년 고안
- 단순 퍼셉트론(Simple Perceptron)
    <img src='./tensorflow/퍼셉트론.png' width='200'>

    - 입력층과 출력층만 구성
    - 입력은 x0, x1, ... xn개의 입력을 가지고  a라는 출력을 가진다 
    - 어떤 입력 기반이 출력으로 확정되는가?
    
- 아이폰 11이 출시했다 구매할것인가?    
    - a(출력) : 구매(1), 비구매(0)
    - 입력 요인은 아이폰 11을 사도 좋은지 관련된 요인을 제시
        - 1) 돈이 감당되는가? (x0)
        - 2) 하드 용량은 충분한가? (x1)
        - 3) 카메라는 만족스러운가? (x2)
        - 4) 무게는 가벼운가? (x3)
        - 5) 지금 쓰는 폰이 문제가 있는가? (x4)
    - 이런 항목중에 과반수가 동의되면 사도 되는가?
        - 부자라면, 1번은 아무 상관없다. 오히려 4번이 문제다 -> 이런 상황마다 다양하게
          가중이 달라진다. 요인들에 대한 무게감(가중치)가 달라진다 
        - 각 입력에 대해서 가중치(W)를 매개변수로 도입한다
        - W0, W1, ...Wn 이라고 한다면
            - 폰이 고장난 사람은 => x4*W4 => x4*5 + x1*1 +,...  가중치를 차등해서 제시
            - 사진찍기가 중요한 사람은 => x2*W2 => x2*5 + x1*2 + ...
    - 이를 기반으로 구입 여부를 검토하는 역치(선택의 기준값)을 C라고 한다면
        if x0*W0 + x1*W1 + ..... xn*Wn > C:
            구매
        else
            비구매
    - 이런 단순 퍼셉트론 하나만으로도 좋은 판단을 구할수 있으나, 여러개를 조합하면 더 복잡한 판단을 할수 있다

#### TensorFlow  

- 구글이 오픈소스로 공개한 머신러닝 라이브러리  
- 딥러닝, 머신러닝에 다양하게 사용된다.   
- 대규모의 수치 연산(계산)을 하는 범용 라이브러리   
- Tensor 는 다차원 행렬 계산을 의미    
- 상용용도가 가능한 라이센스 Apache 2.0   
- 자료가 많다. C++로 코어 개발  
- 전체적으로 계산식을 만들고, 데이터를 넣어서 파이썬 바깥에서 연산을 수행  
- 영상처리시(이미지, 음향) opencv등 병행해서 진행
- www.tensorflow.org
- conda install tensorflow
- 파이썬에서 무거운 작업을 독립적으로 수행하지 않고,
- 상호연관작업을 수행하는 연산간의 그래프(텐서플로우)를 기술하고,
- 그 연산은 파이썬 밖, 텐서플로우 에서 동작한다
