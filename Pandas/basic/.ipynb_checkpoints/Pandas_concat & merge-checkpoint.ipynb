# 데이터프레임 병합

- concat()
- merge()

```python
import numpy as np
import pandas as pd
```

```python
# 데이터 준비
left = pd.DataFrame({
    'key':list('1234'),
    'A'  :['A0','A1','A2','A3'],
    'B'  :['B0','B1','B2','B3']
})

right = pd.DataFrame({
    'key':list('0123'),
    'C'  :['C0','C1','C2','C3'],
    'D'  :['D0','D1','D2','D3'],
})
```

```python
left
```

```python
right
```

## merge()

### 합치기 

- 두개의 데이터프레임을 공통으로 존재하는 컬럼 key 값을 기준으로 merge() 수행
- 교집합의 유사, key 컬럼에 공통된 값을 가진 데이터만 존재시킨다.

```python
pd.merge(left, right, on='key')
```

```python
pd.merge(right, left, on='key')
```

```python
# 교집합이지만 right 기준으로 데이터를 맞춘다.
# 만약 left에만 존재하는 데이터가 있다면 빈데이터는 결측치 처리한다.
# how 옵션으로 한 쪽 데이터를 설정
pd.merge(left, right, on='key', how='right')
```

```python
# 합집합 - outer
# outer옵션 - 공통 되지 않는것은 NaN 처리된다.
pd.merge(left, right, on='key', how='outer')
```

```python
# 교집합 - inner, 생략해도 동일하다.
#inner옵션 -교집합처럼 공통 된 요소를 가진다.
pd.merge(left, right, on='key', how='inner')
```

## concat()

```python
# 상하로 붙였다.(열방향으로 단순한 결합)
pd.concat([left, right], sort=True)
```

```python
# 좌우로 붙인다. (행방향으로 단순한 결합_ + axis=1 )
pd.concat([left, right], sort=True, axis=1)
```
