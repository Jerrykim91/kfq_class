```python
import numpy as np
```

```python
# 0 ~12 사이를 0.01간격으로 알아서 쪼개라
x = np.arange( 0, 12, 0.01 )
x
```

```python
y = np.sin( x )
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
plt.figure( figsize=(10, 6) )
plt.plot(x, y)
# 격자배치
plt.grid()
plt.show()
```

```python
plt.figure( figsize=(10, 6) )
plt.plot(x, y)
plt.grid()
# 라벨
plt.xlabel( 'time' )
plt.ylabel( '값' )
plt.show()
```
