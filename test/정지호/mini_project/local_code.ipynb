```python
import pandas as pd
import numpy as np
```

```python
#시도,시군구 데이터 정리
sigun=pd.read_csv('./data/realdiz/sigun_code.csv', encoding = 'euc-kr')
sido=pd.read_csv('./data/realdiz/sido_code.csv', encoding = 'euc-kr')
```

```python
tmp = []
for n in range(len(sigun)):  ##숫자
    for m in range(len(sido)):
        if  sigun['상위 시도지역코드'][n]==sido['시도 지역코드'][m]:
            tmp.append(  sido['지역명'][m]+' '+ sigun['시군구명'][n] )
        else:
            pass
tmp
            
```

```python
sigun['시군구명'] = tmp
local_code=sigun.set_index('시군구지역코드')
local_code=local_code.drop(['상위 시도지역코드','시군구명'], 1)
```

```python
local_code
```

```python
local_code['지역명'].unique()
```

```python
local_code.to_csv('./data/local_code.csv',sep=',',encoding='utf-8-sig')
```

```python
#눈병 데이터 로딩
eye_sido=pd.read_csv('./data/realdiz/eye_sido.csv', encoding = 'euc-kr')
eye_sigun=pd.read_csv('./data/realdiz/eye_sigun.csv', encoding = 'euc-kr')
#눈병 칼럼명 변경
eye_sido.rename(columns={"시도지역코드":"지역코드"}, inplace = True)
eye_sigun.rename(columns={"시군구지역코드":"지역코드"}, inplace = True)
#눈병 시도,시군 결합
eye = pd.concat([eye_sido,eye_sigun],sort=True)
eye
```

```python
#피부염 데이터 로딩
skin_sido=pd.read_csv('./data/realdiz/skin_sido.csv', encoding = 'euc-kr')
skin_sigun=pd.read_csv('./data/realdiz/skin_sigun.csv', encoding = 'euc-kr')
#피부염 칼럼명 변경
skin_sido.rename(columns={"시도지역코드":"지역코드"}, inplace = True)
skin_sigun.rename(columns={"시군구지역코드":"지역코드"}, inplace = True)
#피부염 시도,시군 결합
skin = pd.concat([skin_sido,skin_sigun],sort=True)
skin
```
