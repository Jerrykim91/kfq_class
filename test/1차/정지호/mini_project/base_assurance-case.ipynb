```python
# 기초생활 수급 인구대비 수급률 엑셀 자료를 읽을 수 있게 파일 경로를 리스트화
down_files = glob('C:/Users/User/Desktop/py_projects/analysis/mini_project/data/base_assur_case/*.csv')
down_files[:2],down_files[-2:]
```

```python
tmp_raws = []
for file in down_files:
    tmp = pd.read_csv( file , encoding='euc-kr')
    tmp_raws.append( tmp )
# 여기는 데이터 확인차!! 
st_raw = pd.concat( tmp_raws )

st_raw
```

```python
st_raw.index = range(len(st_raw))
```

```python
#통계시도명과 시군구명을 처리

#4글자면 : 맨앞과 맨뒤만 쓰기 #경상남도
#5글자이상은 맨앞만    == #3글자면 : 맨앞2글자 절삭 #경기도
#단, 세종은 충남으로

def f_local_code(file_name):
    tmp=[] #'지역구'
    for n in range(len(file_name)):
        if file_name['통계시도명'][n] == '세종특별자치시':
            tmp_f = '충남'
        elif len(file_name['통계시도명'][n]) != 4:
            tmp_f = file_name['통계시도명'][n][0:2]
        else: 
            tmp_f = file_name['통계시도명'][n][0]+file_name['통계시도명'][n][2]
        tmp_s = file_name['통계시군구명'][n]
        tmp.append(tmp_f+' '+tmp_s)
    return tmp
```

```python
tmp= f_local_code(st_raw)
```
