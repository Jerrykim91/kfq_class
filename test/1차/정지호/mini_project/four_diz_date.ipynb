```python
# 감기
cold_sigun=pd.read_csv('./data/realdiz/cold_sigun.csv', encoding = 'euc-kr')
# 칼럼명 변경
cold_sigun.rename(columns={'발생건수(건)':'감기 발생건수'}, inplace = True)
#1401 ~ 1812 까지 만들기 
#201412= 20141231/100
tmp_m = []
for n in range(len(cold_sigun)):
    tmp_m.append(cold_sigun['날짜'][n]//100)
#날짜/ 시군구지역코드/ 감기 발생건수/ 월별
cold_sigun['월별']=tmp_m
#월별만 모으기
cold_m=cold_sigun.drop('날짜',1)
#날짜만 모으기
cold_d=cold_sigun.drop('월별',1)
#월별 지역구별  감기발생건수 합과 평균
cold_m_piv=cold_m.pivot_table(index='월별', columns='시군구지역코드', aggfunc = np.sum)
cold_m_piv
```

```python
# 방법1 함수 돌리기 
def location(diz_m):
    tmp = []
    for n in range(len(local_code)):
        for m in range(len(diz_m.columns.levels[1])):
            if local_code['시군구지역코드'][n] == diz_m.columns.levels[1][m]:
                tmp.append(local_code['지역명'][n])
            else :
                continue
    return tmp
```
